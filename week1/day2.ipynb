## First - let's talk about the Chat Completions API

1. The simplest way to call an LLM
2. It's called Chat Completions because it's saying: "here is a conversation, please predict what should come next"
3. The Chat Completions API was invented by OpenAI, but it's so popular that everybody uses it!

### We will start by calling OpenAI again - but don't worry non-OpenAI people, your time is coming!


````python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')
base_url = os.getenv("OPENAI_BASE_URL", "https://api.openai.com/v1").rstrip("/")
chat_url = f"{base_url}/chat/completions"
if not api_key:
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
else:
    print("you have the codex api from company relax")
````

## Do you know what an Endpoint is?

If not, please review the Technical Foundations guide in the guides folder

And, here is an endpoint that might interest you...

````python
import requests

headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}

payload = {
    "model": "gpt-5-nano",
    "messages": [
        {"role": "user", "content": "Tell me a fun fact"}]
}

payload
````

````python
response = requests.post(
    chat_url,
    headers=headers,
    json=payload
)

data =response.json()
content = data["choices"][0]["message"]["content"]
print(content)
````

# What is the openai package?

It's known as a Python Client Library.

It's nothing more than a wrapper around making this exact call to the http endpoint.

It just allows you to work with nice Python code instead of messing around with janky json objects.

But that's it. It's open-source and lightweight. Some people think it contains OpenAI model code - it doesn't!


````python
# Create OpenAI client

from openai import OpenAI
openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5-nano", messages=[{"role": "user", "content": "Tell me a fun fact about Data scientists who cannot code and fromtend js developers or the fullstack js developers who are now learning AI engineering to save their jobs, but the DS who cannot code are on the green side of field even if they dont know stuff give in pointers"}])

response.choices[0].message.content

````

## And then this great thing happened:

OpenAI's Chat Completions API was so popular, that the other model providers created endpoints that are identical.

They are known as the "OpenAI Compatible Endpoints".

For example, google made one here: https://generativelanguage.googleapis.com/v1beta/openai/

And OpenAI decided to be kind: they said, hey, you can just use the same client library that we made for GPT. We'll allow you to specify a different endpoint URL and a different key, to use another provider.

So you can use:

```python
gemini = OpenAI(base_url="https://generativelanguage.googleapis.com/v1beta/openai/", api_key="AIz....")
gemini.chat.completions.create(...)
```

And to be clear - even though OpenAI is in the code, we're only using this lightweight python client library to call the endpoint - there's no OpenAI model involved here.

If you're confused, please review Guide 9 in the Guides folder!

````python
"""
Week 2 exercise: Technical Q&A with Gradio + streaming + system prompt + model switch
+ optional calculator tool (OpenAI path only; see note below).
Run:  python week2_technical_tutor.py
Requires: pip install gradio openai python-dotenv
Ollama:  ollama serve  &&  ollama pull llama3.2
"""
import ast
import json
import operator
import os
import gradio as gr
from dotenv import load_dotenv
from openai import OpenAI
load_dotenv(override=True)
# --- System prompt (expertise / behavior) ---
SYSTEM_MESSAGE = """You are a patient senior software engineer helping students in an LLM engineering course.
Explain technical topics clearly. Use short sections, bullets, and small code examples when useful.
If the user pastes code, explain what it does line-by-line when appropriate."""
# --- Clients ---
OPENAI_BASE_URL = os.getenv("OPENAI_BASE_URL", "https://api.openai.com/v1")
cloud = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=OPENAI_BASE_URL.rstrip("/"),
)
ollama = OpenAI(
    base_url=os.getenv("OLLAMA_BASE_URL", "http://localhost:11434/v1").rstrip("/"),
    api_key="ollama",
)
# Map dropdown labels -> (client, model_id, supports_remote_tools)
MODEL_CHOICES = {
    "GPT-4.1 mini (OpenAI)": (cloud, "gpt-4.1-mini", True),
    "Llama 3.2 (Ollama)": (ollama, "llama3.2", False),
}
TOOLS = [
    {
        "type": "function",
        "function": {
            "name": "calculator",
            "description": "Evaluate a simple arithmetic expression with + - * / and parentheses. Numbers only.",
            "parameters": {
                "type": "object",
                "properties": {
                    "expression": {
                        "type": "string",
                        "description": 'e.g. "(2 + 3) * 4"',
                    }
                },
                "required": ["expression"],
            },
        },
    }
]
def safe_calc(expression: str) -> str:
    """Very small safe-ish math: only numbers and + - * / ( )."""
    allowed_ops = {
        ast.Add: operator.add,
        ast.Sub: operator.sub,
        ast.Mult: operator.mul,
        ast.Div: operator.truediv,
        ast.USub: operator.neg,
    }
    def _eval(node):
        if isinstance(node, ast.Constant) and isinstance(node.value, (int, float)):
            return node.value
        if isinstance(node, ast.BinOp) and type(node.op) in allowed_ops:
            return allowed_ops[type(node.op)](_eval(node.left), _eval(node.right))
        if isinstance(node, ast.UnaryOp) and isinstance(node.op, ast.USub):
            return -_eval(node.operand)
        raise ValueError("Unsupported expression")
    tree = ast.parse(expression, mode="eval")
    return str(_eval(tree.body))
def run_openai_with_optional_tool(prompt: str, use_tool: bool):
    """
    Non-streaming path when tools are enabled: one model turn + optional tool call + final answer.
    Streaming + tools in one loop is longer; this satisfies the 'tool' requirement clearly.
    """
    messages = [
        {"role": "system", "content": SYSTEM_MESSAGE},
        {"role": "user", "content": prompt},
    ]
    kwargs = dict(
        model="gpt-4.1-mini",
        messages=messages,
    )
    if use_tool:
        kwargs["tools"] = TOOLS
        kwargs["tool_choice"] = "auto"
    reply = cloud.chat.completions.create(**kwargs)
    msg = reply.choices[0].message
    # Tool branch
    if use_tool and msg.tool_calls:
        tc = msg.tool_calls[0]
        name = tc.function.name
        args = json.loads(tc.function.arguments or "{}")
        messages.append(
            {
                "role": "assistant",
                "content": msg.content,
                "tool_calls": [
                    {
                        "id": tc.id,
                        "type": "function",
                        "function": {"name": name, "arguments": tc.function.arguments},
                    }
                ],
            }
        )
        if name == "calculator":
            result = safe_calc(args.get("expression", ""))
            messages.append(
                {
                    "role": "tool",
                    "tool_call_id": tc.id,
                    "content": result,
                }
            )
        final = cloud.chat.completions.create(
            model="gpt-4.1-mini",
            messages=messages,
        )
        return final.choices[0].message.content or ""
    return msg.content or ""
def stream_response(prompt: str, model_label: str, use_tool: bool):
    """Streaming for the main chat; tool mode uses non-stream final text (see homework bonus)."""
    if not prompt.strip():
        yield ""
        return
    client, model_id, can_tool = MODEL_CHOICES[model_label]
    # Bonus: tool path only on OpenAI cloud (Ollama tool support varies by model/setup)
    if use_tool and can_tool:
        text = run_openai_with_optional_tool(prompt, use_tool=True)
        # Fake stream by chunks so UI still feels responsive
        out = ""
        step = max(1, len(text) // 40)
        for i in range(0, len(text), step):
            out = text[: i + step]
            yield out
        return
    messages = [
        {"role": "system", "content": SYSTEM_MESSAGE},
        {"role": "user", "content": prompt},
    ]
    stream = client.chat.completions.create(
        model=model_id,
        messages=messages,
        stream=True,
    )
    result = ""
    for chunk in stream:
        piece = chunk.choices[0].delta.content or ""
        result += piece
        yield result
def build_ui():
    model_dropdown = gr.Dropdown(
        choices=list(MODEL_CHOICES.keys()),
        value="GPT-4.1 mini (OpenAI)",
        label="Model",
    )
    tool_checkbox = gr.Checkbox(
        label="Use calculator tool (OpenAI only; uses function calling then answers)",
        value=False,
    )
    demo = gr.Interface(
        fn=stream_response,
        title="Week 2 — Technical tutor",
        description="Streaming technical Q&A. Toggle tool mode for a simple calculator tool on the OpenAI path.",
        inputs=[
            gr.Textbox(label="Your technical question", lines=8),
            model_dropdown,
            tool_checkbox,
        ],
        outputs=gr.Markdown(label="Answer"),
        examples=[
            ["What does `yield from` do in Python?", "GPT-4.1 mini (OpenAI)", False],
            ["Compute (128 + 256) / 8", "GPT-4.1 mini (OpenAI)", True],
        ],
        flagging_mode="never",
    )
    return demo
if __name__ == "__main__":
    build_ui().launch()
    
````
